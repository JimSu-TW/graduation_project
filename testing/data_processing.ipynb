```python
import pandas as pd
import re
```

```python
def strQ2B(ustring):
    """把字串全形轉半形"""
    ss = []
    for s in ustring:
        rstring = ""
        for uchar in s:
            inside_code = ord(uchar)
            if inside_code == 12288:  # 全形空格直接轉換
                inside_code = 32
            elif (inside_code >= 65281 and inside_code <= 65374):  # 全形字元（除空格）根據關係轉化
                inside_code -= 65248
            rstring += chr(inside_code)
        ss.append(rstring)
    return ''.join(ss)
```

```python
raw_content = pd.read_csv('../training_data/Content_cut.csv')
phone_list = pd.read_csv('../db_related/db_csv/tbl_product.csv')['name'].tolist()
brand_list = pd.read_csv('../db_related/db_csv/tbl_brand.csv')['Brand'].tolist()
source_list = pd.read_csv('../db_related/db_csv/tbl_source.csv')['Source'].tolist()
with open('../useful_data/stopwords_tc.txt', encoding='utf-8', mode='r') as f:
    stop_words = []
    for l in f:
        stop_words.append(l.strip())
```

```python
raw_content_list = []
for index, row in raw_content.iterrows():
    if(index<1):
        temp_list = row['content_cut']
        raw_content_list.extend(eval(temp_list))
```

```python
raw_content_list_to_half = []
process_content = []
delete_list = phone_list + brand_list +source_list
lower_delete_list = []
count = 0
for w in delete_list:
    lower_delete_list.append(w.lower())
for word in raw_content_list:
    word_to_half = strQ2B(word)
    raw_content_list_to_half.append(word_to_half)
for word_half in raw_content_list_to_half:
    flag = False
    temp_word = word_half.strip()
    temp_word = temp_word.lower()
    temp_word = temp_word.replace(" ", "")
    if temp_word in stop_words:
        continue
    elif (temp_word=="「" or temp_word=="」"):
        continue
    elif (temp_word.find('http') != -1):
        continue
    elif (temp_word.find('ios') != -1 or temp_word.find('android') != -1):
        continue
    elif(temp_word.find('▼') !=-1 or temp_word.find('-') !=-1 or temp_word.find(':') !=-1 or temp_word.find('♥') !=-1 or temp_word.find('\\') !=-1 or temp_word.find('%') !=-1 or temp_word.find('&') !=-1 or temp_word.find('{') !=-1 or temp_word.find('/')!=-1 or temp_word.find(',')!=-1 or temp_word.find("'")!=-1):
        continue
    elif(temp_word.find('se3se2') !=-1):
        continue
    else:
        for delete_word in lower_delete_list:
            temp_delete_word = delete_word.replace(" ","")
            if(temp_delete_word.find(temp_word)!=-1):
                flag = True
                break
        if(flag == False):
            count += 1
            print("count: " + str(count) + " word: "+ temp_word)
            process_content.append(temp_word)
```

```python
with open('../training_data/process_content.txt', encoding='utf-8', mode='w') as f:
    for word in process_content:
        f.write(word + "\n")
```
