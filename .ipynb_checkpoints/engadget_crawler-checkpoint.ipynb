```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from datetime import date
from bs4 import BeautifulSoup
import sys
import time
import pandas as pd
```

```python
#new chrome driver

from selenium import webdriver

def OpenChromeDriver():
    #開啟瀏覽器新分頁，下載google chrome driver
    ban_notify=webdriver.ChromeOptions()
    prefs={"profile.default_content_setting_values.notifications" : 2}
    ban_notify.add_experimental_option("prefs",prefs)
    gchrome=webdriver.Chrome(options=ban_notify)
    return gchrome

def FinishedCrawler(driver):
    driver.close()
```

```python
#抓取貼文連結

def getPostLink(url,driver):
    driver.get(url)
    post_link_list = []
    soup = BeautifulSoup(driver.page_source,"html.parser")
    for l in soup.find_all("a", class_="C(engadgetFontBlack) Fw(400) Td(n) Bdbw(1px):h Bdbc(lightgrey):h Bdbs(s):h Fz(30px) Lh(35px) Fz(22px)--sm Lh(25px)--sm"):
        temp = []
        temp.append(l.get_text())
        temp.append("https://chinese.engadget.com/"+l["href"])
        post_link_list.append(temp)
    link_df = pd.DataFrame(post_link_list)
    link_df = link_df.rename(columns={0: 'Post Title', 1: 'Link'})
    FinishedCrawler(driver)
    return link_df
```

```python
#抓取貼文留言

def getPostComment(url,driver):
    driver.get(url)
            
    #Get Comment       
    soup = BeautifulSoup(driver.page_source,"html.parser")
    content_all = str(soup.find("div",id="post-center-col"))
    content_html = content_all[:content_all.find('<div id="post-bottom-social-share">')]
    content_fix_soup = BeautifulSoup(content_html, "html.parser")
    content_text_all=[]
    for i in content_fix_soup.find_all("p"):
        if(i.find("a") != None):
            continue
        else:
            content_text_all.append(i.get_text())
    FinishedCrawler(driver)
    return content_text_all
```

```python
temp_df=getPostLink("https://chinese.engadget.com/tag/iphone%2013%20pro",OpenChromeDriver())
```

```python
contents = []
for link in temp_df.loc[:,"Link"]:
    contents.append(getPostComment(link,OpenChromeDriver()))
    
content_df = pd.DataFrame({'Content': contents})
temp_df = temp_df.join(content_df)
temp_df
```

```python
import datetime
now = datetime.datetime.now().strftime("%Y%m%d")
file_name = "craw_data/Engadget_iphone13_Data"+now+".csv"
temp_df.to_csv(file_name, index=False, encoding="utf-8")
```
