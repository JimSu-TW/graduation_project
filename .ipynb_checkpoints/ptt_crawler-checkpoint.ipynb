```python
#new chrome driver

from selenium import webdriver

def OpenChromeDriver():
    #開啟瀏覽器新分頁，下載google chrome driver
    ban_notify=webdriver.ChromeOptions()
    prefs={"profile.default_content_setting_values.notifications" : 2}
    ban_notify.add_experimental_option("prefs",prefs)
    gchrome=webdriver.Chrome(options=ban_notify)
    return gchrome

def FinishedCrawler(driver):
    driver.close()
```

```python
#ptt get comment

from selenium import webdriver
from bs4 import BeautifulSoup
import requests 

def getPttComment(URL):
    
    my_headers = {'cookie': 'over18=1;'}
    response = requests.get(URL, headers = my_headers)
    
#     Get Comment       
    soup = BeautifulSoup(response.text,"html.parser")
    articles = soup.find_all('div', 'push')
    
    comment_text=[]
    for article in articles:
        messages = article.find('span','f3 push-content').getText().replace(':','').strip()
        comment_text.append(messages)
#     列表轉換為字串
    comment_text = "。".join(comment_text)
    return comment_text
    FinishedCrawler(driver)
```

```python
def get_one_page(URL,driver):

    post_title_list = []
    post_link_list = []
    
    my_headers = {'cookie': 'over18=1;'}
    response = requests.get(URL, headers = my_headers)
    
    soup = bs4.BeautifulSoup(response.text,"html.parser")
    titles = soup.find_all('div','title')
    
    for t in titles:
        temp1 = []
        # 爬標題
        temp1.append(t.text.strip())
        post_title_list.append(temp1)
                        
    for items in soup.select('div[class = "title"] a'):
        temp2 = []
        # 爬網址
        temp2.append("https://www.ptt.cc" + str(items.get('href')))
        post_link_list.append(temp2)
        
    Title_list = pd.DataFrame(post_title_list)
    Title_list = Title_list.rename(columns={0: 'Post Title'})
    
    Link_list = pd.DataFrame(post_link_list)
    # 合併    
    df_ptt_TNL = pd.concat([Title_list, Link_list],axis=1)
    df_ptt_TNL = df_ptt_TNL.rename(columns={0: 'Link'})

    return df_ptt_TNL
```

```python
import pandas as pd

# 把多頁的資訊加入
storing = pd.DataFrame(columns=['Post Title', 'Link'])

# 第一頁網址不同
first_page = get_one_page('https://www.ptt.cc/bbs/iOS/search?q=iphone+13', OpenChromeDriver())
storing = storing.append(first_page)
```

```python
# 多頁爬取
# 含第一頁先爬個5面

import bs4
import time
import pandas as pd

start = 2 # 設定起始網頁 
# 最舊 = 14q
number = 14    # 設定要從開始頁面往後爬多少個

for i in range(start,number+1):
    
    # 組成 正確 URL
    link = "https://www.ptt.cc/bbs/iOS/search?page="+str(i)+"&q=iphone+13"
    
    # 執行單頁面網頁爬蟲
    Title_Link = get_one_page(link, OpenChromeDriver())
    storing = storing.append(Title_Link, ignore_index=True)
    time.sleep(1)
storing    

# ptt iphone 13網址
# 首頁    https://www.ptt.cc/bbs/iOS/search?q=iphone+13
# 後面分頁    https://www.ptt.cc/bbs/iOS/search?page=2&q=iphone+13
```

```python
# 先把爬下來的主題跟網址存起來
storing.to_csv('PPT_iphone13_title_link_0603.csv', index=False, encoding="utf-8")
```
